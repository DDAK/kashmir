```python
import folium
```

```python
m = folium.Map(
    location=[20.5937, 78.9629],
    tiles=None,
    zoom_start=5  # Limited levels of zoom for free Mapbox tiles.
)
```

```python
def foo(m):
    if m['properties']['st_nm'] == 'Ladakh':
        return {
            'fillColor': 'red',
            'color': 'red',
            'fillOpacity': 0.3
        } 
    else:
        return {
            'fillColor': 'cyan',
            'color': 'cyan',
            'fillOpacity': 0.3
        }
```

```python
folium.GeoJson(
    'kashmir_ladakh_siachen.json',
    style_function=foo
).add_to(m)

```
